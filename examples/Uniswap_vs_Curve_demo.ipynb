# Messari Python Package Example

The purpose of this jupyter notebook is to provide an example of what can be done using the Messari Python package. In this example we will be using APIs from Messari and DeFi Llama to do some in depth analysis of everybodies two favorite AMMs, Uniswap and Curve. Both of these APIs are easy to interface with using the Messari Package so let's see what we can discover!

## Setup

```python
import matplotlib.pyplot as plt
import pandas as pd
```

## Price data

```python
from messari.timeseries import get_metric_timeseries

metric = 'price'
assets = ['uni', 'crv']
start = '2021-01-01'
end = '2021-11-01'
price_timeseries_df = get_metric_timeseries(asset_slugs=assets, asset_metric=metric, start=start, end=end)
```

```python
price_timeseries_df['uni']['close'].plot()
plt.show()
```

```python
price_timeseries_df['uni']['close'].plot()
plt.show()
```

```python
price_timeseries_df.xs('close', axis=1, level=1).plot()
plt.show()
```

```python
metric = 'mcap.realized'
mcap_timeseries_df = get_metric_timeseries(asset_slugs=assets, asset_metric=metric, start=start, end=end)
```

```python
mcap_timeseries_df.plot()
plt.show()
```
