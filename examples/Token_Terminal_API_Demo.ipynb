# Token Terminal

```python
from messari.tokenterminal import TokenTerminal
API_KEY='Your API Key'
tt = TokenTerminal(api_key=API_KEY)
protocols = ['uniswap', 'compound']
start_date = "2021-10-01"
end_date = "2021-10-10"
```

## get_project_ids

Returns all project ids available in Token Terminal

```python
project_ids = tt.get_project_ids()
print(project_ids)
```

## get_all_protocol_data

Returns an overview of latest data for all projects, ranging from metadata such as launch dates, logos brand colors and Twitter followers to more fundamental metrics such as Revenue, GMV, TVL and P/S ratios.

```python
all_protocol_data = tt.get_all_protocol_data()
all_protocol_data
```

## get_protocol_data

Returns a time series of the latest data for a given project, ranging from metadata such as Twitter followers to more fundamental metrics such as Revenue, GMV, TVL and P/S ratios.

```python
protocol_data = tt.get_protocol_data(protocols)
protocol_data
```

## get_historical_metric_data

Returns the time series of a specified metric for a given list of project.

```python
historical_mktcap = tt.get_historical_metric_data(protocols, metric="market_cap")
historical_mktcap
```
